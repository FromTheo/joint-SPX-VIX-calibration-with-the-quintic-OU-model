### $K^\text{frac}$ kernel 

```python
t_vals = np.linspace(0, 2, 500)
decomposition = get_optimal_decomposition("RECORD_QF.txt", 60)
T = t_vals[-1]
```

```python
H = 0.5
trajectoires_0_5, weights_0_5 = generate_X_quantized_trajectories_frac(T, t_vals, decomposition, H)
```

```python
H = 0.05
trajectoires_0_1, weights_0_1 = generate_X_quantized_trajectories_frac(T, t_vals, decomposition, H)
```

```python
fig, ax = plt.subplots(1, 2, figsize =(14,5))

for i, traj in enumerate(trajectoires_0_5) : 
    ax[0].plot(t_vals, traj)
    ax[0].set_title(r"$\text{Quantizer } \widehat X_t,\;H = 0.5$", fontsize=14)

for i, traj in enumerate(trajectoires_0_1) : 
    ax[1].plot(t_vals, traj)
    ax[1].set_title(r"$\text{Quantizer } \widehat X_t,\;H = 0.05$", fontsize=14)

for i in range(2) : 
    ax[i].set_xlabel(r"$\text{Time }t$", fontsize=12)
    ax[i].tick_params(axis='both', labelsize=12)
    ax[i].xaxis.set_major_formatter(lambda x, _: f"${x:g}$")
    ax[i].yaxis.set_major_formatter(lambda y, _: f"${y:g}$")
    
plt.tight_layout()
plt.show()
```

### $K^\text{exp}$ kernel

```python
t_vals = np.linspace(0, 0.5, 500)
eps = 1/52
decomposition = get_optimal_decomposition("RECORD_QF.txt", 60)
T = t_vals[-1]
```

```python
H = 0.5
trajectoires_0_5, weights_0_5 = generate_X_quantized_trajectories_exp(T, t_vals, decomposition, H, eps, file_loader = load_quantization_gaussian)
```

```python
H = -0.1
trajectoires_0_1, weights_0_1 = generate_X_quantized_trajectories_exp(T, t_vals, decomposition, H, eps, file_loader = load_quantization_gaussian)
```

```python
fig, ax = plt.subplots(1, 2, figsize =(14,5))

for i, traj in enumerate(trajectoires_0_5) : 
    ax[0].plot(t_vals, traj)
    ax[0].set_title(r"$\text{Quantizer } \widehat X_t,\;H =0.5$", fontsize=14)

for i, traj in enumerate(trajectoires_0_1) : 
    ax[1].plot(t_vals, traj)
    ax[1].set_title(r"$\text{Quantizer } \widehat X_t,\;H = -0.1$", fontsize=14)

for i in range(2) : 
    ax[i].set_xlabel(r"$\text{Time }t$", fontsize=12)
    ax[i].tick_params(axis='both', labelsize=12)
    ax[i].xaxis.set_major_formatter(lambda x, _: f"${x:g}$")
    ax[i].yaxis.set_major_formatter(lambda y, _: f"${y:g}$")

plt.tight_layout()
plt.show()
```
